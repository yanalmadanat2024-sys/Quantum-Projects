```python
from qiskit import QuantumCircuit
import numpy as np

# Create a 3-qubit quantum circuit
qc = QuantumCircuit(3)

# Step 1: Apply Hadamard gate on the first qubit
qc.h(0)

# Step 2: Apply controlled phase rotation between qubit 1 and qubit 0
qc.cp(np.pi/2, 1, 0)  # R_2 gate (π/2 phase shift)

# Step 3: Apply controlled phase rotation between qubit 2 and qubit 0
qc.cp(np.pi/4, 2, 0)  # R_3 gate (π/4 phase shift)

# Step 4: Apply Hadamard gate on the second qubit
qc.h(1)

# Step 5: Apply controlled phase rotation between qubit 2 and qubit 1
qc.cp(np.pi/2, 2, 1)  # R_2 gate (π/2 phase shift)

# Step 6: Apply Hadamard gate on the third qubit
qc.h(2)

# Step 7: Swap qubits 0 and 2 to reverse order (final step in QFT)
qc.swap(0, 2)

# Draw the circuit
qc.draw("mpl")
# This code constructs a 3-qubit Quantum Fourier Transform (QFT) circuit using Qiskit.
```
